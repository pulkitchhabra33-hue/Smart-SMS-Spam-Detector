```python
## import necessary libraries
import pandas as pd
import numpy as np
```

```python
## Importing the dataset
messages= pd.read_csv(r"C:\Users\pulki\Smart SMS Spam Detector\data\spam.csv", encoding= 'Latin-1')
messages.head()
```

```python
# Modifying the dataset
messages= messages[['v1', 'v2']]
messages.columns= ['label', 'message']
messages.head()
```

```python
messages['message'].iloc[354]
```

```python
# Data Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
# Text Preprocessing
# (remove symbols, lowercase, stemming, stopwords)
corpus= []
for i in range(0, len(messages)):
    review= re.sub('[^a-zA-Z]', " ", messages['message'][i])
    review= review.lower()
    review= review.split()
    review= [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review= " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Feature Extraction - CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features= 5000)
X= cv.fit_transform(corpus).toarray()
```

```python
# Label Encoding
y= pd.get_dummies(messages['label'])
y= y.iloc[:, 1].values
y
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.2, random_state= 12)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model= MultinomialNB().fit(X_train, y_train)
```

```python
# Prediction
y_pred= spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer= WordNetLemmatizer()
```

```python
corpus= []
for i in range(0, len(messages)):
    review= re.sub('[^a-zA-Z]', " ", messages['message'][i])
    review= review.lower()
    review= review.split()
    review= [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review= " ".join(review)
    corpus.append(review)
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words= []
for sent in corpus:
    sents= sent_tokenize(sent)
    for s in sents:
        words.append(simple_preprocess(s))
```

```python
import gensim
```

```python
model= gensim.models.Word2Vec(words, window= 5, min_count= 2)
```

```python
def avg_word2vec(doc):
    vectors=  [model.wv[word] for word in doc.split() if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis= 0)
```
